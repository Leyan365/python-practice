# finding the 2nd and  3rd largest value

```python
# finding the 2nd and  3rd largest value
#  without sorting

nums = [23,46,67,44,547,23,6,75]

first = second = third = float('-inf')

for n in nums:
    if n > first:
        third = second
        second = first
        first = n
    elif n > second and n != first:
        third = second
        second = n
    elif n > third and n != second and n != first:
        third = n

print('2nd largest:', second)
print('3rd largest:', third)
print()

# with sorting

unique_nums = sorted(list(set(nums)) , reverse=True)

print('2nd largest:', unique_nums[1])
print('3rd largest:', unique_nums[2])
print()

# using heapq
import heapq

largest_three = heapq.nlargest(3 , unique_nums)
second = largest_three[1]
third = largest_three[2]

print('2nd largest:',second)
print('3rd largest:', third)
```

```python
# top 5 largest using heapq
import heapq
k = 5
min_heap = []

for n in nums:
    if len(min_heap) < k:
        heapq.heappush(min_heap ,n)

    elif n > min_heap[0]:
        heapq.heapreplace(min_heap,n)

print('Top 5 largest:', sorted(min_heap , reverse=True))
```

```python
# finding the 2nd and 3rd smallest value


nums = [2,3,5,6,7,8,5,43,54,64,7568,568,43346,146,5]


# with sorting
n_nums = sorted(list(set(nums)))

print('2nd smallest', n_nums[1])
print('3rd smallest', n_nums[2])

# using heapq
import heapq

heap_list = heapq.nsmallest(3, n_nums)

print('2nd smallest', heap_list[1])
print('3rd smallest', heap_list[2])

# wihtout sorting

first = second = third = float('inf')

for n in nums:
    if n < first:
        third = second
        second = first
        first = n
    elif n < second and n != first:
        third = second
        second = n
    elif n < third and n != first and n != second:
        third = n

print('2nd smallest', second)
print('3rd smallest', third)
```

```python
# finding the top 5 smallest values using heapq

import heapq

k = 5
max_heap = []

for n in n_nums:
    if len(max_heap) < k:
        heapq.heappush(max_heap,-n)
    
    elif n < -max_heap[0]:
        heapq.heapreplace(max_heap,-n)

# convert values back 
max_heap = sorted([-x for x in max_heap])

print('Top 5 smallest' , max_heap)
```
