```python
nums = [1, 2, 5, 73, 7, 3, 1, 4, 8, 9, 45, 4, 56, 9]
nums = list(set(nums))
```

```python
def bubble_sort(arr):
    for i in range(len(arr)):
        for j in range(0, len(arr)-i-1):
            if arr[j] > arr[j+1]:
                arr[j] , arr[j+1] = arr[j+1] , arr[j]
    return arr
```

```python
def linear__search(arr,target):
    for i , value in enumerate(arr):
        if value == target:
            return i
    return -1
```

```python
def binary_search(arr,target):
    low = 0
    high = len(arr)-1

    while low <= high:
        mid = (low + high) // 2

        if arr[mid] == target:
            return mid
        elif arr[mid] < target:
            low = mid + 1
        else:
            high = mid - 1
    return -1
```

```python
def factorial(n):
    if n == 0 or n == 1:
        return 1
    return n * factorial(n-1)

def fibonacci(n):
    a = 0
    b = 1
    for _ in range(n):
        a , b = b , a + b
    return a
```

```python
def Check_Prime(n):
    if n <= 1 : return False
    if n == 2 : return True
    if n % 2 == 0 : return False
    for i in range (3, int(n ** 0.5) + 1 , 2):
        if n % i == 0 :
            return False
    return True
```

```python
def reverse_string(s):
    return s[::-1]

def palindrome(s):
    return s == s[::-1]
```

```python
print("Bubble Sort:", bubble_sort(nums))
print("Linear Search (3):", linear__search(nums, 3))
print("Binary Search (5):", binary_search(nums, 5))
print("Factorial (5):", factorial(5))
print("Fibonacci (4):", fibonacci(4))
print("Check Prime (7):", Check_Prime(7))
print("Reverse String ('hello'):", reverse_string('hello'))
print("Palindrome ('cake'):", palindrome('cake'))
```

```python
def Fibonacci(n):
    if n == 0:
        return 0
    elif n == 1:
        return 1
    else:
        return Fibonacci(n-1) + Fibonacci(n-2)

for i in range(0,10):
    print(Fibonacci(i), end = " ")  
```
