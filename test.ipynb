## What Are Getters and Setters?

A getter is a method used to retrieve (get) the value of a private variable.

Think:
"Tell me whats inside, but dont let me touch it."

###  Setter

A setter is a method used to change (set) the value of a private variable safely.

Think:
"You can update it, but only through this controlled method."

# What is Inheritance? (Simple Definition)

Inheritance lets one class (child) use the properties and methods of another class (parent).

### It helps with:

``` 
Code reuse

Reducing repetition

Creating logical relationships
```

#### Example:
A Dog and Cat are both Animals.

````python
# Basic Syntax

class Parent:
    pass

class Child(Parent):
    pass
````

````python
class Animal:
    def __init__(self,name):
        self.name = name

    def eat(self):
        return f"{self.name} is eating."
    
# Animal is the parent/base class

# It provides:

# name attribute

# eat() method

# Both Dog and Cat will automatically inherit these.

class Dog(Animal):

    def sound(self):
        return "bark"

# Dog is a child/derived class

# It inherits everything from Animal

# It adds its own sound() method

class Cat(Animal):

    def sound(self):
        return 'meow'
    
# Cat is another child/derived class

# Same inheritance

# Different sound() behavior (polymorphism)
````

````python
d = Dog("Buddy")
c = Cat("Mimi")

print(d.eat())     # inherited from Animal
print(d.sound())   # Dogs own method

print(c.eat())     # inherited from Animal
print(c.sound())   # Cats own method
````

# What is Method Overriding? (Simple Definition)

Method overriding happens when a child class provides its own version of a method that already exists in the parent class.

The childs method replaces the parents method at runtime.

### You override a method when:
```
You want the child class to behave differently

But still share the same method name and signature
```

````python
class AnimalNew:
    def __init__(self, name):
        self.name = name

    def sound(self):
        return "Some Animal Sound"
    
class DogNew(AnimalNew):
    def sound(self):
        return 'bark'
    
class CatNew(AnimalNew):
    def sound(self):
        return 'meow'

d = DogNew("Buddy")
print(d.sound())
print(d.name)
````

# What Is Abstraction? (Simple Explanation)

Abstraction hides complex details and only shows the essential features.

### In Python, abstraction is done using abstract classes.
```
You define a base class with methods that must be implemented

Child classes must override those methods

You cannot create objects of an abstract class 
```

````python
from abc import ABC, abstractmethod

class Animal(ABC):  # abstract class

    @abstractmethod
    def sound(self):
        pass  # no implementation , only the idea

    def eat(self):
        return "This animal is eating."
    
# child class must implement the abstract method

class Dog(Animal):
    def sound(self):
        return "bark"

class Cat(Animal):
    def sound(self):
        return "meow"
    
d = Dog()
c = Cat()

print(d.sound())
print(c.sound())
print(d.eat())
````
